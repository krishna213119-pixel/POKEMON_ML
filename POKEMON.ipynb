```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df1[['Winner']] = sc.fit_transform(df1[['Winner']])
```

```python
pokemon = pd.read_csv("pokemon.csv")
combats = pd.read_csv("combats.csv")

m1 = combats.merge(pokemon, left_on="First_pokemon", right_on="#", how="left")
m1 = m1.rename(columns={c: f"First_{c}" for c in pokemon.columns if c != "#"}).drop(columns=["#"])

m2 = m1.merge(pokemon, left_on="Second_pokemon", right_on="#", how="left")
m2 = m2.rename(columns={c: f"Second_{c}" for c in pokemon.columns if c != "#"}).drop(columns=["#"])
```

```python
m2.head()
```

```python
y = (m2["Winner"] == m2["First_pokemon"]).astype(int)
print(y.value_counts(normalize=True))
```

```python
m2.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
m2.columns
```

```python
x = m2[['First_HP','First_Attack','First_Defense','First_Sp. Atk','First_Sp. Def','First_Speed',
        'Second_HP','Second_Attack','Second_Defense','Second_Sp. Atk','Second_Sp. Def','Second_Speed']]

y = (m2['Winner'] == m2['First_pokemon']).astype(int)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from joblib import dump

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

model12 = RandomForestClassifier(n_estimators=100, random_state=42)
model12.fit(x_train, y_train)

print("X features:", x.shape[1])
print("Trained model expects:", model12.n_features_in_)

# overwrite the old file
dump(model12, "model.joblib")
print("✅ Saved NEW model.joblib")
y_pred=model12.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
acc=accuracy_score(y_test, y_pred)
acc
```
